```python
import moleculegraph
import numpy as np
import toml

import unittest
```

```python
mdma = "[CH3][C][b2][HN][CH3][C][C1][C2][C3][C4][b3][O1][C][O2][r5][C5][C6][r6]" 

molecule = moleculegraph.molecule(mdma)
print(np.array([molecule.atom_numbers.astype(float)]) )
print( molecule.distance_matrix )
#print( molecule.bond_matrix )
print( molecule.bond_list )
molecule.visualize(saveto="mdma.pdf")
print(molecule.angle_keys)
print(molecule.torsion_keys)
print(molecule.bond_keys)
```

```python
molecule.neighbour_list, molecule.idx_neighbour_list
```

```python

mdma_expected = ['[C][CH3]', '[C][HN]', '[C][C]', '[CH3][HN]', '[C1][C]',
       '[C1][C2]', '[C1][C6]', '[C2][C3]', '[C3][C4]', '[C3][O2]',
       '[C4][O1]', '[C4][C5]', '[C][O1]', '[C][O2]', '[C5][C6]']

tc = unittest.TestCase()
tc.assertListEqual( sorted(list(molecule.bond_keys)), sorted(list(mdma_expected)), "lists not equal" ) 
```

```python
import networkx as nx

options = {
    "node_size": 1200,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 5,
    "width": 3,
    "with_labels":True,
    "alpha":.3,
    "font_size":12,
}

graph  = nx.Graph()
for b0,b1 in molecule.bond_list:
    graph.add_edge(b0,b1)

labels = {}

for i, name in enumerate(molecule.atom_names):
    labels[i] = name
    
pos = nx.kamada_kawai_layout(graph)

nx.draw_networkx(graph, pos,labels=labels,**options)
nx.draw_networkx_labels(graph, pos , labels, font_size=12, font_color="black")
```
